# Encoding

When organizing and analyzing data in statistics, **nominal** and **ordinal** data are two fundamental types of categorical (non-numerical) data.

The easiest way to tell them apart is to look at whether the **order** of the categories matters.

## 1. Nominal Data (Labels Only)

Nominal data is used to name, label, or classify things into distinct groups. The categories have **no inherent order, ranking, or numerical value**. One category isn't "higher" or "better" than another; they are just different.

- **Key Characteristic:** You cannot sort them in a meaningful logical order (other than alphabetically).
- **Examples:**
- **Eye color:** Brown, Blue, Green, Hazel.
- **Marital status:** Single, Married, Divorced, Widowed.
- **Car brands:** Toyota, Ford, BMW, Tesla.
- **Binary outcomes:** Yes / No, Pass / Fail.

## 2. Ordinal Data (Labels with a Order)

Ordinal data also classifies things into groups, but the categories have a **clear, natural order or ranking**. While we know which category is "more" or "less" than another, we cannot precisely measure the exact *distance* between them.

- **Key Characteristic:** The sequence matters, but the mathematical difference between rankings isn't equal or measurable. (For example, the difference between 1st and 2nd place in a race might be 0.1 seconds, but between 2nd and 3rd it might be 5 seconds).
- **Examples:**
- **Satisfaction ratings:** Dissatisfied, Neutral, Satisfied.
- **Socioeconomic status:** Low-income, Middle-income, High-income.
- **Competition results:** 1st place, 2nd place, 3rd place.
- **Education level:** High School, Bachelor's, Master's, PhD.

## Quick Comparison Summary

| Feature | Nominal Data | Ordinal Data |
| --- | --- | --- |
| **Has Categories?** | Yes | Yes |
| **Ordered/Ranked?** | No | **Yes** |
| **Mathematical Value?** | No | No |
| **Best Measure of Center** | Mode (most frequent) | Median (middle value) or Mode |
| **Core Purpose** | To identify/group | To rank/scale |

In machine learning and data science, **nominal encoding** (often called categorical encoding) is the process of converting categorical variables—like city names, colors, or job titles—into numerical values so that mathematical algorithms can process them.

Because nominal data has no inherent order, how you choose to encode it plays a massive role in model performance.

## Pros and Cons of Nominal Encoding Techniques

There isn't just one way to encode nominal data. The pros and cons depend entirely on the method you choose. The two most common techniques are **One-Hot Encoding** and **Label/Ordinal Encoding** (though using Label Encoding on nominal data is a common trap).

### 1. One-Hot Encoding (The Standard Approach)

This creates a new binary column (0 or 1) for every unique category in your dataset.

- **Pros:**
- **No Fake Hierarchy:** It correctly treats all categories as equally different. The model won't mistakenly think "Blue" is greater than "Red."
- **Works Great with Linear Models:** Algorithms like Linear Regression or Logistic Regression handle these explicit binary boundaries very well.
- **Cons:**
- **The Curse of Dimensionality:** If a column has 100 unique categories (e.g., US States), One-Hot Encoding turns that 1 column into 50 columns. This explodes your dataset size, slows down training, and can cause overfitting.
- **Multicorrelation (The Dummy Variable Trap):** The columns become perfectly predictable from one another, which can break certain statistical models.

### 2. Label Encoding (The Risky Approach)

This simply assigns an arbitrary number to each category (e.g., Red = 1, Blue = 2, Green = 3).

- **Pros:**
- **Space Efficient:** It keeps your dataset memory-light by keeping everything in a single column.
- **Cons:**
- **Introduces Fake Order:** This is the biggest flaw. A machine learning model will look at the numbers and assume $3 \gt 1$ (Green is greater than Red), or that the "average" of Red and Green is Blue $(1 + 3) / 2 = 2$. For nominal data, this is mathematically nonsensical and can ruin your model's accuracy.

## The Crucial Statistical Concept: The Dummy Variable Trap

If you use One-Hot Encoding, you must understand a vital statistical concept called **Multicollinearity**—specifically, the **Dummy Variable Trap**.

### What is Multicollinearity?

Multicollinearity occurs when two or more predictor variables in a statistical model are highly correlated, meaning one can be linearly predicted from the others with a high degree of accuracy.

### How One-Hot Encoding Causes It

Imagine a nominal variable for "Gender" with two categories: `Male` and `Female`. If you One-Hot Encode this, you get two columns:

1. `Is_Male` (1 or 0)
2. `Is_Female` (1 or 0)

If a row has a `1` for `Is_Male`, it *must* have a `0` for `Is_Female`. They are perfectly inversely correlated. Statistically, the second column provides **zero new information**.

Mathematically, this causes severe issues for models like Ordinary Least Squares (OLS) Linear Regression because it makes the design matrix non-invertible. In plain English: **the algorithm's math breaks down**, leading to unstable, wildly fluctuating coefficients.

### The Statistical Solution: $N-1$ Encoding

To avoid this trap, you should always drop one of the encoded columns. If a feature has $N$ unique categories, you only keep $N-1$ dummy columns.

- If `Is_Male` is `0`, the model automatically infers that the person is female. The dropped category becomes the **reference/baseline group** against which all other categories are statistically measured.

You should use **nominal encoding** whenever your dataset contains categorical variables that have **no inherent ranking, order, or numerical distance** between them, and you need to feed this data into a machine learning algorithm.

Because machine learning models are essentially massive calculators that perform matrix multiplication, they cannot understand the words "Red," "Blue," or "Green"—they require numbers.

However, choosing *which* nominal encoding technique to use depends entirely on the **cardinality** of your data and the **type of algorithm** you are running.

## 1. When to Use One-Hot Encoding

One-Hot Encoding is the default, go-to choice for nominal data under specific conditions.

- **Low-to-Medium Cardinality:** Use it when your nominal feature has a small number of unique categories (e.g., less than 15 or 20). Examples include `Gender` (Male, Female, Non-binary) or `Continent` (Asia, Europe, Africa, etc.).
- **When Using Linear Models or SVMs:** Algorithms like Linear Regression, Logistic Regression, and Support Vector Machines rely on linear decision boundaries. One-Hot Encoding creates a clear, distinct binary geometric space for each category, which these models handle perfectly.
- **When Using Neural Networks:** Deep learning models can easily process the sparse, high-dimensional matrices created by One-Hot Encoding.

## 2. When to Use Target Encoding (or Frequency Encoding)

If you have **high-cardinality** nominal data, One-Hot Encoding will break your model by creating thousands of columns.

- **High Cardinality:** Use this when a nominal variable has dozens, hundreds, or thousands of unique categories. Examples include `Zip Code`, `IP Address`, `Device Model`, or `Job Title`.
- **When Using Tree-Based Algorithms:** Gradient boosting trees (like XGBoost, LightGBM, and CatBoost) or Random Forests excel at handling target-encoded variables because they can easily split the data along the continuous, numeric averages that target encoding provides.

## 3. When to Use Count / Frequency Encoding

This approach replaces each category with the number of times (or percentage of times) it appears in the dataset.

- **When Frequency Itself is a Signal:** Use this when the popularity or rarity of a nominal category tells you something important about the target. For example, in fraud detection, an IP address or credit card country that appears 10,000 times in an hour is highly suspicious compared to one that appears once.

## Decision Matrix: Which Encoding to Use?

| If your nominal data has... | And your ML model is... | **Your best encoding choice is...** |
| --- | --- | --- |
| **Low Cardinality** (e.g., $\le 10$ categories) | Any Model (Linear, Trees, NN) | **One-Hot Encoding** (Drop 1 column for linear models) |
| **High Cardinality** (e.g., $100+$ categories) | Tree-Based (XGBoost, Random Forest) | **Target Encoding** (with smoothing) |
| **High Cardinality** (e.g., $100+$ categories) | Linear Model or Neural Network | **Feature Embedding** (Entity Embeddings) or **Hash Encoding** |
| **Varying Cardinality** | Any Model (Fraud/Anomalies) | **Count / Frequency Encoding** |

## Related Statistical Concept: The Curse of Dimensionality

A critical statistical concept dictating *when* to use which encoding is the **Curse of Dimensionality**.

As you add more features (columns) to a dataset, the volume of the data space grows exponentially, making the available data points incredibly **sparse**.

If you use One-Hot Encoding on a high-cardinality nominal variable (like 500 unique product IDs), you suddenly inject 500 columns of mostly zeros into your dataset. In this massive, sparse high-dimensional space, data points become equidistant from one another.

As a result:

1. **Distance-based algorithms** (like K-Nearest Neighbors or K-Means Clustering) completely fail because they can no longer accurately calculate which points are genuinely "close" to one another.
2. **Models easily overfit**, finding random noise patterns in the sea of zeros that won't generalize to new data.

This statistical reality is exactly why data scientists transition away from One-Hot Encoding and toward Target Encoding or Embeddings as category counts grow.

#### 3. OHE (One Hot Encoding)

```python
# One-Hot Encoding using pandas.get_dummies -> produces binary indicator columns for each category
col = "MSZoning"
print("Before:", df[col].unique())

# Create the one-hot columns (DataFrame) using pandas utility
mszoning_ohe_pandas = pd.get_dummies(df[col], prefix="MSZoning")
# Concatenate with original column for a compact display and drop duplicates to show one row per category
pandas_comparison = pd.concat([df[col], mszoning_ohe_pandas], axis=1)
unique_pandas_ohe = pandas_comparison.drop_duplicates(subset=[col]).reset_index(drop=True)
# Output: DataFrame showing each original category and its one-hot indicator columns
unique_pandas_ohe
```

```python
# Method 2: sklearn OneHotEncoder (produces numpy array or sparse matrix depending on config)
from sklearn.preprocessing import OneHotEncoder

# Configure encoder: `sparse_output=False` returns dense numpy array; `drop=None` keeps all categories
# dense numpy array will contain all the zeros and ones for the one-hot encoding of each category. This will require more memory for large datasets, but is easier to work with for small datasets.
ohe = OneHotEncoder(sparse_output=False, drop=None)
# fit_transform returns array shape (n_samples, n_categories)
mszoning_ohe_sklearn = ohe.fit_transform(df[[col]])
print("Sklearn OneHotEncoder:", mszoning_ohe_sklearn)
# Convert array to DataFrame for display with feature names generated by the encoder
ohe_df = pd.DataFrame(mszoning_ohe_sklearn, columns=ohe.get_feature_names_out([col]))
# Concatenate and deduplicate to show mapping from category to one-hot columns
sklearn_comparison = pd.concat([df[col], ohe_df], axis=1)
unique_sklearn_ohe = sklearn_comparison.drop_duplicates(subset=[col]).reset_index(drop=True)
# Output: DataFrame showing one-hot encoded representation per original category
unique_sklearn_ohe
```

#### 4. Target Encoding

**Target Encoding** (also known as Mean Encoding) is a highly effective technique for converting nominal categorical variables into numerical values. Instead of creating new columns or assigning arbitrary numbers, Target Encoding replaces each category with the **expected value of the target variable** for that specific category.

For a binary classification problem (e.g., predicting if a customer will churn: 1 for Yes, 0 for No), a category is replaced by the **mean of the target variable** (the probability of churning) for that category.

## How It Works: A Quick Example

Imagine you are predicting house prices (Target) based on the City (Nominal Feature).

| City (Feature) | House Price (Target) | Target Encoded City |
| --- | --- | --- |
| New York | $500,000 | **$450,000** |
| New York | $400,000 | **$450,000** |
| Miami | $300,000 | **$250,000** |
| Miami | $200,000 | **$250,000** |

- The average house price in New York is $450,000. Every instance of "New York" is replaced with `450000`.
- The average house price in Miami is $250,000. Every instance of "Miami" is replaced with `250000`.

## Pros and Cons of Target Encoding

### Pros:

- **No Dimensionality Explosion:** Unlike One-Hot Encoding, Target Encoding always outputs exactly **one column**, regardless of whether your category has 5 or 5,000 unique values. This saves massive amounts of memory and computational time.
- **Captures Strong Relationships:** It directly injects the relationship between the feature and the target into the dataset, which often gives tree-based models (like XGBoost or LightGBM) a massive performance boost.

### Cons:

- **Severe Risk of Overfitting:** Because this method uses the target variable to encode the feature, it leaks information about the target into the training data. The model can easily memorize these encodings rather than learning generalized patterns.
- **The "Rare Category" Problem:** If a category appears only once in your dataset (e.g., a tiny town with one mansion priced at $2,000,000), Target Encoding will assign it a value of `2000000`. This completely distorts the data based on a single outlier.

## When to Use Target Encoding

You should opt for Target Encoding when:

1. You have **high-cardinality nominal features** (columns with dozens, hundreds, or thousands of unique categories, like zip codes, IP addresses, or product IDs) where One-Hot Encoding would create too many columns.
2. You are using **tree-based machine learning algorithms** (Gradient Boosting, Random Forests) which can exploit the monotonic relationship created by target encoding beautifully.

## Crucial Statistical Concept: Empirical Bayes Smoothing

To fix the "Rare Category" problem and prevent drastic overfitting, statisticians and data scientists use a technique called **Smoothing** (strongly rooted in **Empirical Bayes estimation**).

Instead of relying solely on the category's specific average, smoothing blends the category average with the **global average** of the entire dataset.

### The Math Behind It

The smoothed value for a category is calculated using a weighted average:

$$
\hat{x}_i = \lambda(n_i) \cdot \mu_i + (1 - \lambda(n_i)) \cdot \mu_{\text{global}}
$$

Where:

- $n_i$ is the number of times category $i$ appears in the dataset.
- $\mu_i$ is the raw target mean for category $i$.
- $\mu_{\text{global}}$ is the global target mean across the entire dataset.
- $\lambda(n_i)$ is a weight function (a value between 0 and 1) that depends on the sample size $n_i$.

### How It Solves the Problem

- **For a common category (high $n_i$):** $\lambda$ is close to 1. The model trusts the category's specific mean ($\mu_i$) because there is plenty of data to back it up.
- **For a rare category (low $n_i$):** $\lambda$ is close to 0. The formula shrinks the encoded value toward the **global mean** ($\mu_{\text{global}}$). If that tiny town only has one data point, the encoding pulls its value back down toward the average house price of the whole country, preventing the model from making wild assumptions based on a single sample.

#### 5. Frequency Encoding

**Frequency Encoding** (also known as Count Encoding) is a straightforward and effective technique for handling nominal categorical variables. Instead of creating new binary columns or relying on the target variable, it replaces each category with the number of times it appears in the dataset (its count) or the percentage of its occurrence (its frequency).

## How It Works: A Quick Example

Imagine a column for `Car Brand` in a dataset of 100 vehicles:

| Car Brand (Feature) | Count in Dataset | Frequency Encoding (Count) | Frequency Encoding (Percentage) |
| --- | --- | --- | --- |
| Toyota | 60 times | **60** | **0.60** |
| Ford | 30 times | **30** | **0.30** |
| Tesla | 10 times | **10** | **0.10** |

Every time "Toyota" appears in your rows, it is simply replaced by `60` (or `0.60`).

## Pros and Cons of Frequency Encoding

### Pros:

- **No Dimensionality Explosion:** Just like Target Encoding, Frequency Encoding keeps your data memory-light by outputting exactly **one column** per categorical feature, avoiding the thousands of sparse columns that One-Hot Encoding might produce.
- **Blends Group Sizes Intuitively:** It preserves a very specific piece of information: how common or rare a category is.
- **Computationally Efficient:** It only requires a basic frequency mapping step (a single pass to count the occurrences), making it incredibly fast to compute even on massive datasets.

### Cons:

- **The "Collision" Problem:** If two entirely different categories appear the exact same number of times in your dataset, they will get the exact same encoded value. For example, if "BMW" also appears exactly 10 times, both Tesla and BMW will be encoded as `10`. The model loses the ability to tell them apart.
- **Ignores the Target Variable:** Unlike Target Encoding, Frequency Encoding doesn't look at what you are trying to predict. If a category's popularity has no correlation with the target variable, this encoding inserts useless information into the model.

## When to Use Frequency Encoding

You should choose Frequency Encoding when:

1. **You have high-cardinality nominal features** (like Zip Codes, User IDs, or Product Categories) where One-Hot Encoding is out of the question.
2. **Frequency itself is a strong predictor.** This is incredibly common in specific domains:
- **Fraud Detection:** If an IP address or Credit Card Country appears 50,000 times in an hour, its high frequency is a massive red flag.
- **Web Analytics / E-commerce:** Highly frequent search terms or product views behave differently than rare, "long-tail" search items.
3. **You are using tree-based machine learning algorithms** (like Random Forests or Gradient Boosting) because they can easily create splits based on popularity (e.g., "Is this product category in the top 10% of popularity?").

## Related Statistical Concepts

### 1. Power Law / Zipf's Law (The Long Tail)

In many real-world datasets, categorical frequencies do not follow a uniform or normal distribution. Instead, they follow a **Power Law distribution** (often generalized as Zipf's Law or the 80/20 rule).

This statistical principle states that a small number of categories will appear an overwhelming majority of the time, while a massive number of categories will appear very rarely (known as the "Long Tail").

Frequency encoding maps this phenomenon into a numerical scale. By converting categories into their frequencies, the encoder groups the massive sea of rare "Long Tail" attributes into a clustered numeric space (near 0 or 1, depending on count vs. percentage), allowing your machine learning model to easily treat rare categories as a collective group rather than learning individual, noisy patterns for each one.

### 2. Frequency vs. Probability (The Law of Large Numbers)

When you use the percentage variation of Frequency Encoding, you are calculating the **empirical probability** (or relative frequency) of that category occurring within your sample space:

$$
P(\text{Category}) = \frac{\text{Count of Category}}{\text{Total Number of Rows}}
$$

According to the **Law of Large Numbers**, as your dataset grows larger, this sample relative frequency will converge closer to the true, theoretical probability of that category existing in the real world. This makes the frequency representation highly stable and reliable when training on large production-scale datasets, though it can remain volatile and prone to noise if your training sample size is too small.

```python
# Frequency encoding example: inspect categories before computing frequencies
col = "Neighborhood"
print("Before:", df[col].unique())
# Output: numpy array of unique neighborhood names
```

```python
# Method 1: create a frequency mapping using pandas.value_counts
# Output: dictionary mapping category -> observed count
freq_encoding = df[col].value_counts().to_dict()
```

```python
# Show the frequency mapping dictionary (output: python dict)
freq_encoding
```

```python
# Map each category to its frequency count and store as new column (output: numeric column)
df["Neighborhood_freq"] = df[col].map(freq_encoding)
```
