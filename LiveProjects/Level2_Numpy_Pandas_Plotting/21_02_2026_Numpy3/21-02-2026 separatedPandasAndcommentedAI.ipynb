## Array Operations {#array-operations}

| Operation | Syntax | Purpose | Returns |
|-----------|--------|---------|---------|
| **Element-wise Addition** | `arr1 + arr2` | Add corresponding elements | Array |
| **Element-wise Multiplication** | `arr1 * arr2` | Multiply corresponding elements | Array |
| **Dot Product (1D)** | `np.dot(arr1, arr2)` | Multiply + sum elements | Scalar |
| **Matrix Multiplication (2D)** | `np.dot(mat1, mat2.T)` or `mat1 @ mat2.T` | Linear algebra multiplication | Matrix |
| **Transpose** | `arr.T` | Swap rows and columns | Array |

## Aggregation Functions {#aggregation-functions}

| Function | Syntax | Purpose | Returns |
|----------|--------|---------|---------|
| **sum()** | `np.sum(arr, axis=None)` | Sum all/axis elements | Scalar/Array |
| **min()** | `np.min(arr)` | Find minimum value | Scalar |
| **max()** | `np.max(arr)` | Find maximum value | Scalar |
| **argmin()** | `np.argmin(arr)` | Index of minimum | Index |
| **argmax()** | `np.argmax(arr)` | Index of maximum | Index |
| **mean()** | `np.mean(arr, axis=None)` | Average of elements | Scalar/Array |
| **std()** | `np.std(arr, axis=None)` | Standard deviation | Scalar/Array |
| **var()** | `np.var(arr, axis=None)` | Variance | Scalar/Array |
| **cumsum()** | `np.cumsum(arr)` | Cumulative sum | Array |
| **median()** | `np.median(arr, axis=None)` | Median value | Scalar/Array |
| **percentile()** | `np.percentile(arr, q)` | q-th percentile | Scalar/Array |

**Axis Parameter**: axis=0 (sum rows, keep cols), axis=1 (sum cols, keep rows), axis=None (global)

## Sort/Search/Count Functions {#sort-search-count}

| **argsort()** | `np.argsort(arr)` | Indices that would sort array | Index array |
| **searchsorted()** | `np.searchsorted(sorted_arr, val)` | Find insertion position | Index |
| **unique()** | `np.unique(arr)` | Find unique elements | Unique array |
| **bincount()** | `np.bincount(arr)` | Count frequency of each value | Count array |

## Array Iteration {#iteration}

| Method | Syntax | Purpose | Returns |
|--------|--------|---------|---------|
| **for loop (1D)** | `for x in arr:` | Iterate through elements | Element |
| **for loop (2D)** | `for row in arr_2d:` | Iterate through rows | Row (1D array) |
| **nditer()** | `for x in np.nditer(arr):` | Iterate all elements (any dims) | Scalar element |

## Copy vs View vs Reference {#copy-view-reference}

| Concept | Syntax | Memory | Changes Reflected | Use Case |
|---------|--------|--------|-------------------|----------|
| **Reference** | `b = a` | Same object, same data | ✅ Both see changes | Aliasing (not recommended) |
| **View** | `b = a.view()` | Different object, same data | ✅ Both see changes | Memory efficient, linked arrays |
| **Copy** | `b = a.copy()` | Different object, different data | ❌ Independent | Safe modifications |

**Check**: `np.shares_memory(a, b)` returns True for view/reference, False for copy

## Stacking Operations {#stacking}

| Function | Syntax | Purpose | Result |
|----------|--------|---------|--------|
| **vstack()** | `np.vstack((arr1, arr2))` | Stack arrays row-wise (vertically) | Taller array |
| **hstack()** | `np.hstack((arr1, arr2))` | Stack arrays column-wise (horizontally) | Wider array |

## String/Character Operations {#string-operations}

| Function | Syntax | Purpose | Returns |

| **np.char.upper()** | `np.char.upper(arr)` | Convert to uppercase | String array |
| **np.char.lower()** | `np.char.lower(arr)` | Convert to lowercase | String array |

| **np.char.split()** | `np.char.split(arr, sep)` | Split string by separator | Array of lists |
| **np.char.find()** | `np.char.find(arr, substring)` | Find index of substring | Index array |
| **np.char.count()** | `np.char.count(arr, substring)` | Count substring occurrences | Count array |

## Key Notes

- **Broadcasting**: Operations automatically expand dimensions when needed
- **In-place operations**: Some methods modify original array (check documentation)
- **Memory efficiency**: Views are faster than copies for large arrays
- **Axis indexing**: Always specify axis for multi-dimensional operations


```python
# Import NumPy library for advanced numerical computing
# Alias 'np' is standard convention - all numpy functions accessed via np.function_name()
import numpy as np
```

```python
# Create two Python lists and demonstrate list concatenation
# In Python, + operator concatenates lists (not element-wise like NumPy)
# Result: [1, 2, 3, 4, 5, 6] - lists are joined end-to-end
a_list = [1,2,3]
b_list = [4,5,6]

print(a_list + b_list)
```

```python
# Create 1D NumPy arrays for element-wise operations
# NumPy enables vectorized operations (faster than Python loops)
# + operator performs element-wise addition (not concatenation like lists)
# Result: [1+4, 2+5, 3+6] = [5, 7, 9]
a = np.array([1,2,3])
b = np.array([4,5,6])
print(a+b)
```

```python
# Element-wise multiplication (Hadamard product) - multiply corresponding elements
# * operator is element-wise, NOT matrix multiplication
# Result: [1*4, 2*5, 3*6] = [4, 10, 18]
# For marix multiplication, the column of first matrix must match the row of second matrix
# Here the arraysa are 1D and the corresponding dimensions are 1*3 and 1*3, so the column of first array is not matching with the row of second array
# Internally what happens in dot product for arrays is, the second array is transposed first and then both the arrays are multiplied 
a*b
```

```python
# Dot product - multiply element-wise THEN sum the results
# np.dot(a, b) = (1*4) + (2*5) + (3*6) = 4 + 10 + 18 = 32
# Used in: machine learning, physics, linear algebra
print(np.dot(a,b))
```

```python
# Create 2D arrays (matrices with 2 rows, 3 columns) for matrix operations
# Both have same shape (2, 3) - needed for element-wise operations
a_1 = np.array([[1,2,3],[4,5,6]])
b_1 = np.array([[1,2,3],[4,5,6]])
```

```python
# Check dimensions of both matrices using .shape attribute
# Returns: tuple (rows, columns) e.g., (2, 3) means 2 rows, 3 columns
# .shape is critical for debugging matrix operations
print(a_1.shape)
print(b_1.shape)
```

```python
# Matrix multiplication using np.dot()
# Formula: A (2x3) @ B^T (3x2) = Result (2x2)
# Note: b_1.T transposes - swaps rows/columns so shapes align for multiplication
# (2,3) @ (3,2) = (2,2) valid for multiplication
print(b_1.T) # Check shape of transposed matrix
np.dot(a_1, b_1.T)
```

```python
# Matrix multiplication using @ operator (Python 3.5+)
# @ is shorthand for np.dot() - cleaner syntax for matrix operations
# Equivalent to: np.dot(a_1, b_1.T)
a_1@b_1.T
```

```python
# Real-world example: Store inventory sold per product
# Products: apple, banana, smartphone, toy, furniture
# Data: Sales from 2 states (regions/stores)
# Shape (2,5): 2 states, 5 products - each element = quantity sold
shop_sold_amazon = np.array([
    [100, 200, 20, 30, 5], # state-1-product-sold
    [200, 400, 40, 60, 10] # state-2-product-sold
])
print(shop_sold_amazon.shape)
```

```python
arr = np.array([3, 6, 1, 1, 1, 1, 8, 2, 5, 6])
```

```python
# Sort array and get sorted indices
# np.sort(arr) - returns sorted values
# np.argsort(arr) - returns INDICES that would sort the array
# argsort() is powerful: use returned indices to sort other arrays the same way
print("Sorted:", np.sort(arr))
print("Sorted indices:", np.argsort(arr))
# Instead of returning the sorted values themselves, 
# it returns an array of indices that, when used to index the original array, would produce a sorted sequence.
```

```python
# Search, unique, and count operations
# np.searchsorted() - find insertion position for value in sorted array
# np.unique() - get all unique (distinct) elements from array
# np.bincount() - count frequency of each integer from 0 to max value
# Returns count array where index = value, element = count
print("Insert 4 position:", np.searchsorted(np.sort(arr), 4))
print("Unique elements:", np.unique(arr))
print("Count of each element:", np.bincount(arr)) # MIN - 0 , maX - 8

# 0,1,2,3,4,5,6,7,8
# 0,4,1....
```

```python
# Create a reference assignment - NOT a copy
# b = a means: b and a point to THE SAME object in memory
# Changes to either will affect both (no data safety)
a = np.array([1, 2, 3])
b = a
```

```python
# Create a deep copy - independent array with copied data
# b = a.copy() creates a NEW array with COPIED data
# Changes to b don't affect a, and vice versa
# Memory usage: Higher (two copies of data), but safe
b = a.copy()
```

```python
print(id(a))
print(id(b))
```

```python
b[0] = 100
```

```python
print(a)
```

```python
b = a.copy()
```

```python
print(id(a))
print(id(b))
```

```python
# Create a view - different Python object, SAME underlying data
# b = a.view() creates a NEW object that references SAME data as a
# Changes to values affect both objects (data is shared)
# Memory efficient: No data copy, but objects are linked
a = np.array([1, 2, 3,])

b = a.view()
```

```python
print(id(a))
print(id(b))
```

```python
print(np.shares_memory(a, b))
```

```python
# Stacking operations - combine multiple arrays into single array
# vstack (vertical stack) - stack arrays on top of each other (increase rows)
# hstack (horizontal stack) - stack arrays side by side (increase columns)

a = np.array([[1,2]])
b = np.array([[3,4]])
```

```python
print(np.vstack((a,b)))
```

```python
print(np.hstack((a,b)))
```

# numpy char and strings

```python
# np.char.<function_name>()
```

```python
# String addition using np.char.add() - concatenates strings
# Different from numeric addition (which adds numbers)
# Combines each element from arr1 with corresponding element from arr2
# Result: ['hello_user', 'world_123']
a = np.array(['hello', 'world'])
b = np.array(['_user', '_123'])

print(np.char.add(a, b))
```

```python
# Repeat/multiply strings - repeats each string n times
# np.char.multiply(arr, n) where n is repetition count
# Result: ['hellohello', 'worldworld'] (each string repeated 2 times)
print(np.char.multiply(a, 2))
```

```python
# Case conversion operations on string arrays
# np.char.upper() - convert all to UPPERCASE
# np.char.lower() - convert all to lowercase
# np.char.capitalize() - capitalize first letter only
a = np.array(['hello', 'WORLD'])

print(np.char.upper(a)) 
print(np.char.lower(a)) 
print(np.char.capitalize(a))
```
