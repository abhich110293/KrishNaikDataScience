## Data Loading & Inspection {#loading-inspection}

| Function | Syntax | Purpose | Returns |
|----------|--------|---------|---------|
| **read_csv()** | `pd.read_csv(url/path)` | Load CSV file into DataFrame | DataFrame |
| **head()** | `df.head(n)` | Display first n rows (default: 5) | DataFrame subset |
| **tail()** | `df.tail(n)` | Display last n rows (default: 5) | DataFrame subset |
| **info()** | `df.info()` | Show column names, types, memory usage | None (prints info) |
| **describe()** | `df.describe()` | Statistical summary of numeric columns | DataFrame |

## DataFrame Information {#dataframe-info}

| Property/Method | Syntax | Purpose | Returns |
|-----------------|--------|---------|---------|
| **shape** | `df.shape` | Get (rows, columns) dimensions | Tuple: (int, int) |
| **dtypes** | `df.dtypes` | Get data types of all columns | Series |
| **columns** | `df.columns` | Get column names | Index |
| **index** | `df.index` | Get row indices | RangeIndex or Index |
| **type()** | `type(df)` | Get object type | Class name |

## Accessing Data {#accessing-data}

| Method | Syntax | Purpose | Returns |
|--------|--------|---------|---------|
| **Single Column** | `df['column_name']` | Access one column | Series |
| **Multiple Columns** | `df[['col1', 'col2']]` | Access multiple columns | DataFrame |
| **By Position** | `df.iloc[0]` | Access by row number | Series |
| **By Label** | `df.loc[0]` | Access by index label | Series |
| **Boolean Indexing** | `df[df['col'] > value]` | Filter rows by condition | DataFrame |

## Common Aggregations {#aggregations}

| Function | Syntax | Purpose | Returns |
|----------|--------|---------|---------|
| **sum()** | `df.sum()` or `df['col'].sum()` | Sum all numeric values | Scalar or Series |
| **mean()** | `df.mean()` | Calculate average | Scalar or Series |
| **min()** | `df.min()` | Find minimum value | Scalar or Series |
| **max()** | `df.max()` | Find maximum value | Scalar or Series |
| **count()** | `df.count()` | Count non-null values | Scalar or Series |
| **groupby()** | `df.groupby('col').agg(func)` | Group and aggregate | DataFrame/Series |

## Key Notes

- **Series**: 1D labeled array (like a column)
- **DataFrame**: 2D labeled table (like a spreadsheet)
- **Index**: Row labels (default: 0, 1, 2, ...)
- **Columns**: Column names/labels
- **View vs Copy**: Selecting columns returns a view; modifying requires `.copy()`

```python
# Import the pandas library and assign alias 'pd'
# By convention, pandas is always imported as 'pd' for brevity
# All pandas functions/classes are now accessed with pd.function_name()
import pandas as pd
```

```python
# Store the URL to the Titanic dataset hosted on GitHub
# This is a public dataset commonly used for learning data analysis
# Includes passenger info: Survived, Pclass, Name, Sex, Age, Fare, etc.
titanic_dataset_github_url = "https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv"
```

```python
# Read CSV file from URL and load into a DataFrame
# pd.read_csv() is the most common function for loading CSV files
# Returns a DataFrame object - a 2D table with rows and columns
# df = variable name for the DataFrame (common convention)
df = pd.read_csv(titanic_dataset_github_url)
```

```python
# Check the type/class of the variable 'df'
# Returns: pandas.core.frame.DataFrame
# This confirms df is a DataFrame object (the main data structure in pandas)
type(df)
```

```python
# Display first 5 rows of the DataFrame (default behavior)
# Useful for quickly inspecting data structure and contents
# Returns: DataFrame with 5 rows and all columns
# Use df.head(n) to display first n rows instead
df.head()
```

```python
# Display last 5 rows of the DataFrame (default behavior)
# Opposite of head() - useful for checking the end of the dataset
# Returns: DataFrame with last 5 rows and all columns
# Use df.tail(n) to display last n rows instead
df.tail()
```

```python
# Display first 10 rows of the DataFrame
# Parameters: n (number of rows to display)
# Returns: DataFrame with 10 rows showing structure and sample data
df.head(10)
```

```python
# Access a single column by name using bracket notation
# Returns a Series - a 1D labeled array (single column)
# Syntax: df['column_name'] or df.column_name (if no spaces in name)
# Series includes index (row numbers) and values (data)
df['Name']
```

```python
# Access multiple columns by passing a list of column names
# Note: Double brackets [[ ]] - outer brackets for selection, inner for list
# Returns: DataFrame with selected columns (2D, not 1D Series)
# Syntax: df[['col1', 'col2', 'col3']] for multiple columns
df[['Name', 'Age']]
```

```python
# Get the dimensions (shape) of the DataFrame
# Returns: Tuple (rows, columns) e.g., (891, 12) = 891 rows, 12 columns
# Useful for understanding dataset size before processing
# Syntax: df.shape - returns tuple, df.shape[0] = rows, df.shape[1] = columns
df.shape
```

```python
# List all attributes and methods available for the DataFrame object
# Returns: List of all accessible properties and functions
# Useful for discovering available methods (head, tail, shape, sum, mean, etc.)
# Can also use: help(df), df.info(), or IDE autocomplete for exploration
dir(df)
```
